```python
import pandas as pd
from bs4 import BeautifulSoup
import re

# loading the data 
df = pd.read_csv("/Users/sarthaktyagi/Desktop/SHLassesment/SHLscrapedData.csv")  

# text cleaning function -> remove html tagsand remove white spaceas
def clean_text(text):
    if pd.isna(text):
        return ""
    #html tags
    text = BeautifulSoup(text, "html.parser").get_text()

    # Remove extra whitespace
    text = re.sub(r"\s+", " ", text).strip()
    return text

df["Description"] = df["Description"].apply(clean_text)
df["Assessment Name"] = df["Assessment Name"].apply(clean_text)

# Normalize test_type 
df["Test Type"] = df["Test Type"].apply(lambda x: clean_text(x).upper())

# Normalize remote_support / adaptive_support to Yes/No
def normalize_flag(value):
    value = str(value).lower()
    if "yes" in value:
        return "Yes"
    elif "no" in value:
        return "No"
    else:
        return "Unknown"

df["Remote Support (Yes/No)"] = df["Remote Support (Yes/No)"].apply(normalize_flag)
df["Adaptive/IRT (Yes/No)"] = df["Adaptive/IRT (Yes/No)"].apply(normalize_flag)

# Clean duration â†’ keep only number
def clean_duration(d):
    d = str(d)
    match = re.search(r"\d+", d)
    return int(match.group()) if match else None

df["Duration"] = df["Duration"].apply(clean_duration)

# saving the cleaned csv
df.to_csv("shl_cleaned.csv", index=False)
print("Cleaning complete! Saved as shl_cleaned.csv")
```
